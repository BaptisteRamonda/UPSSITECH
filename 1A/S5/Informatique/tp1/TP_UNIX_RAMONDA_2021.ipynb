35-	Exécuter le script ./boucle.csh
Que se passe-t-il ? Peut-on utiliser la commande kill dans ce cas ? Quelles sont les différentes méthodes permettant d’interrompre, de supprimer ou de suspendre ce processus. Les essayer toutes et relancer au besoin la commande ./boucle.csh

    Il est impossible d'exécuter la commande kill depuis le même terminal sans le symbole "&" à la fin. On a donc 2 moyens pour exécuter la commande kill sur un processus : un si on a toujours le terminal ouvert et un si on ne l'a plus.
